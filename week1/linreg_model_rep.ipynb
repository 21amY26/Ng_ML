```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0,2.0])
y_train = np.array([300.,500.])
print("x_train:", x_train)
print("y_train:", y_train)
```

```python
m = x_train.shape[0]
# m = len(x_train) also works
print(f"No. of training samples: {m}")
```

plotting points using **scatter()** in matplotlib

```python
plt.scatter(x_train, y_train, marker = 'x', c='r')
plt.title("Housing Prices")
plt.xlabel("Sq. Footage (1000ft)")
plt.ylabel("Price (1000$)")
plt.show()
```

```python
w = b = 100
print(f"w: {w}\nb: {b}")
```

## Computing model output

```python
def compute_model_ouput(x,w,b):
    '''
    computes the prediction of the linear regression model
    args:
        x (ndarray(m,)): data, m examples
        w,b (scalar)   : model params
    returns:
        f_wb (ndarray(m,)): model output for each example
    '''
    global m
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b
    return f_wb

```

```python
fwb = compute_model_ouput(x_train, w, b)
#plotting prediction
plt.plot(x_train, fwb, c = 'b',label = 'First Prediction')
# plotting data points
plt.scatter(x_train, y_train, marker = 'x', c='r', label='Actual Values')
plt.title("Housing Prices")
plt.xlabel("Sq. Footage (1000ft)")
plt.ylabel("Price (1000$)")
plt.legend()
plt.show()
```

```python
# trying w=200 and b=100
fwb = compute_model_ouput(x_train, 200, 100)
#plotting prediction
plt.plot(x_train, fwb, c = 'b',label = 'Second Prediction')
# plotting data points
plt.scatter(x_train, y_train, marker = 'x', c='r', label='Actual Values')
plt.title("Housing Prices")
plt.xlabel("Sq. Footage (1000ft)")
plt.ylabel("Price (1000$)")
plt.legend()
plt.show()
```
